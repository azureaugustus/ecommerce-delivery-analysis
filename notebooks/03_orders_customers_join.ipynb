```python
## Orders + Customers Join
# This notebook combines order and customer data to create a final analytical dataset.
# The gao is to enrich order records with customer attributes for easier analysis in SQL.
```

```python
import pandas as pd

customers = pd.read_csv("../data/raw/olist_customers_dataset.csv")
fact_orders = pd.read_csv("../data/clean/fact_orders.csv")
```

```python
fact_orders_enriched = fact_orders.merge(customers[["customer_id", "customer_unique_id"]], on="customer_id",how="left")
```

```python
dim_customers = pd.read_csv("../data/clean/dim_customers.csv")
```

```python
## Joining Data
# Orders are joined with customers using 'customer_unique_id' to combine transactional and dimensional data.
```

```python
final_orders = fact_orders_enriched.merge(dim_customers, on="customer_unique_id",how="left")
```

```python
final_orders.to_csv("../data/clean/fct_orders.csv", index=False)
```
